```python
import pandas as pd
```

```python
data = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
data.head()
```

```python
data.info()
```

```python
vote_mean = data["vote_count"].mean()
```

```python
print(vote_mean)
```

```python
minvote = data["vote_count"].quantile(0.90)
print (minvote)
```

```python
q_movies = data.copy().loc[data['vote_count']>= minvote]
q_movies.shape
```

```python
def weighted_rating(x,minvote=minvote, vote_mean=vote_mean):
  voters = x['vote_count']
  avg_vote = x["vote_average"]
  return(voters/(voters+minvote)*avg_vote) + (minvote/(minvote+voters)* vote_mean)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score',ascending=False)
```
